```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt # general plotting
import cartopy.crs as ccrs # plot on maps, better than the Basemap module
import re
import scipy.interpolate
from matplotlib.colors import ListedColormap
```

```python
ds = xr.open_dataset("https://opendap.nccs.nasa.gov/dods/OSSE/G5NR-Chem/Heracles/12.5km/0.125_deg/inst/inst1_3d_TRCO3_Nv")
ds # same as print(ds) in IPython/Jupyter environment
time_vector = ds["time"].values
```

```python
dr = xr.Dataset()
dr["ozone [ppb]"]  = ds["trco3"]
```

```python
for tick_time in range(len(time_vector)):
    data_surf = dr["ozone [ppb]"].sel(lev = 27,time=time_vector[tick_time]) #8 km; 339 hpa
    data_surf *= 1e9
    lat = data_surf['lat'].values
    lon = data_surf['lon'].values
    raw_data=data_surf.values
    fig = plt.figure(figsize=[20,10])
    ax = plt.axes(projection=ccrs.Robinson())
    ax.coastlines()
    ax.pcolormesh(lon, lat, data_surf, transform=ccrs.PlateCarree())
    data_surf.plot.imshow(x='lon', y='lat',ax=ax,transform=ccrs.PlateCarree())
    name = "images/time" + str(tick_time) + ".png"
    fig.savefig(name, bbox_inches='tight')
```
